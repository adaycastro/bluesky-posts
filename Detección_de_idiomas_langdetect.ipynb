3. Ejecutar la función para crear una nueva columna llamada 'idioma' con el código de la lengua reconocida en el texto de la publicación

```python
from langdetect import detect

# Función para detectar el idioma
def detectar_idioma(texto):
    try:
        return detect(texto)
    except:
        return "error"

df['idioma'] = df['text'].apply(detectar_idioma)

print(df[['text', 'idioma']].head())
```

```python
df.head(50)
```

**Análisis estadístico de los idiomas detectados**

```python
# 1. Contar la frecuencia de cada código de idioma
conteo_idiomas = df['idioma'].value_counts()

# 2. Calcular el porcentaje que representa cada idioma
porcentaje_idiomas = df['idioma'].value_counts(normalize=True) * 100

# 3. Combinar ambos en un solo DataFrame para mejor visualización
analisis_idiomas = pd.DataFrame({
    'Frecuencia': conteo_idiomas,
    'Porcentaje': porcentaje_idiomas
})

# 4. Redondear porcentajes a 2 decimales
analisis_idiomas['Porcentaje'] = analisis_idiomas['Porcentaje'].round(2)

print(analisis_idiomas)
```

```python
import matplotlib.pyplot as plt

# Paso 1: Separar los idiomas >= 2% y < 2%
umbral = 2  # Porcentaje mínimo
mayores = analisis_idiomas[analisis_idiomas['Porcentaje'] >= umbral]
menores = analisis_idiomas[analisis_idiomas['Porcentaje'] < umbral]

# Paso 2: Sumar los menores y añadirlos como 'Otros'
otros = pd.DataFrame({
    'Frecuencia': [menores['Frecuencia'].sum()],
    'Porcentaje': [menores['Porcentaje'].sum()]
}, index=['Otros'])

# Paso 3: Combinar todo
pie_data = pd.concat([mayores, otros])

# Paso 4: Gráfico de pie
colors = plt.cm.Paired.colors
pie_data['Porcentaje'].plot(
    kind='pie',
    labels=pie_data.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors
)

plt.title('Distribución de idiomas')
plt.ylabel('')  # Ocultar eje Y
plt.tight_layout()
plt.show()
```
